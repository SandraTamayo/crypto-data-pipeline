```python
import requests
import pandas as pd

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}

response = requests.get(url, params=params)
data = response.json()

df = pd.DataFrame(data)

df = df[["id", "symbol", "current_price", "market_cap", "total_volume"]]

df.head()
```

```python
df.to_csv("crypto_data.csv", index=False)
```

```python
import duckdb

con = duckdb.connect()

result = con.execute("""
SELECT *
FROM read_csv_auto('crypto_data.csv')
""").fetchdf()

result
```

```python
result = con.execute("""
SELECT
    id,
    symbol,
    current_price,
    market_cap,
    total_volume,
    market_cap / total_volume AS liquidity_ratio
FROM read_csv_auto('crypto_data.csv')
""").fetchdf()

result
```
